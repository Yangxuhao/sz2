```python
def edu2float(edu):#把学历进行量化
    return np.argwhere(X["education"].unique()==edu)[0,0]
```

```python
def occupation2float(occupation):#把职业进行量化
    return np.argwhere(X["occupation"].unique()==occupation)[0,0]
```

```python
X["occupation"]=X["occupation"].map(occupation2float)
X["education"]=X["education"].map(edu2float)
```

```python
X_train=X.iloc[:-1000]#训练数据
X_test=X.iloc[-1000:]#测试数据
Y_train=Y.iloc[:-1000]
Y_test=Y.iloc[-1000:]
```

```python
logic=LogisticRegression(solver='liblinear',multi_class='auto')
logic.fit(X_train,Y_train)
```

```python
Ynew=logic.predict(X_test)
```

```python
result=(Y_test==Ynew)
accuracy=result.sum()/1000
```
